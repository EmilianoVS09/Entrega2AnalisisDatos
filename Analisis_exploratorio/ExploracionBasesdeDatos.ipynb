## Fase 1: Exploración de Bases de Datos

Para esta fase se documentan tres bases de datos de dos tipos diferentes: datos tabulares 
(IBM HR Analytics y Stack Overflow Developer Survey) y datos de imágenes 
(Reptiles and Amphibians). A continuación se presenta la ficha técnica de cada una.

### 1. IBM HR Analytics Employee Attrition & Performance

Link de la base de datos: https://www.kaggle.com/datasets/pavansubhasht/ibm-hr-analytics-attrition-dataset

**Fuente y tipo:** Secundaria (Kaggle). Dataset tabular sintético generado por científicos 
de datos de IBM con fines educativos y de práctica profesional — no proviene de registros 
reales de empleados, sino de un proceso de simulación controlada.

**Características básicas:**
- Registros: 1,470 empleados
- Atributos: 35 variables (26 numéricas, 9 categóricas)
- Tamaño: 227.98 kB

**Nivel de documentación:** Alto. Cada uno de los 35 atributos está claramente definido 
en el diccionario de datos oficial. Tiene alta actividad en Kaggle con múltiples notebooks 
de referencia.

**Posibles aplicaciones:** Predicción de rotación de empleados (churn laboral), análisis 
de factores de satisfacción, segmentación de perfiles de riesgo y optimización de 
estrategias de retención de talento.

### 2. Stack Overflow Annual Developer Survey

Link de la base de datos: https://www.kaggle.com/datasets/snooptosh/stack-overflow-annual-developer-survey

**Fuente y tipo:** Secundaria (Kaggle / Stack Overflow). Dataset tabular basado en la 
encuesta anual que Stack Overflow aplica a su comunidad de desarrolladores a nivel mundial.

**Características básicas:**
- Registros: Más de 80,000 respuestas
- Atributos: Aproximadamente 85 columnas que cubren lenguajes de programación, 
  herramientas, salarios, experiencia y condiciones laborales
- Tamaño: 94.84 MB

**Nivel de documentación:** La documentación oficial de Stack Overflow es completa y 
detallada. Sin embargo, la versión publicada en Kaggle tiene baja actividad (pocas 
discusiones y notebooks de referencia), lo que reduce el soporte comunitario disponible.

**Posibles aplicaciones:** Análisis de tendencias del mercado laboral tecnológico, 
estudio de brechas salariales por región o tecnología, y exploración de perfiles 
profesionales en el sector de desarrollo de software.

### 3. Reptiles and Amphibians Image Dataset

Link de la base de datos: https://www.kaggle.com/datasets/vencerlanz09/reptiles-and-amphibians-image-dataset

**Fuente y tipo:** Secundaria (Kaggle). Dataset de imágenes organizado por clases 
taxonómicas (especies), recopilado para tareas de visión computacional.

**Características básicas:**
- Registros: Aproximadamente 6,045 imágenes en formato .jpg
- Clases: 15 categorías de especies entre reptiles y anfibios
- Tamaño: 141.05 MB
- Las imágenes presentan diversidad de resoluciones y entornos naturales

**Nivel de documentación:** Alto en cuanto a organización del dataset (carpetas por clase 
claramente etiquetadas). Sin embargo, no incluye metadatos adicionales por imagen 
(localización geográfica, condiciones de captura, etc.).

**Posibles aplicaciones:** Entrenamiento de redes neuronales convolucionales (CNN), 
clasificación taxonómica automatizada, proyectos de monitoreo de biodiversidad y 
transferencia de aprendizaje con modelos preentrenados.

## Justificación de Selección: IBM HR Analytics

Con base en los criterios de completitud, relevancia, documentación y manejabilidad, 
se selecciona el dataset de IBM HR Analytics como base de trabajo para las fases 
siguientes. A continuación se sustenta la decisión frente a cada criterio y en 
comparación con las alternativas evaluadas.

**1. Completitud**
El dataset no presenta valores nulos en ninguna de sus 51,450 celdas (1,470 registros × 35 
atributos), lo que elimina la necesidad de estrategias de imputación complejas y permite 
enfocarse en el análisis sin pérdida de información. Los otros dos datasets no ofrecen 
esta garantía: el de Stack Overflow suele tener respuestas parciales por la naturaleza 
voluntaria de la encuesta, y el de imágenes no cuenta con metadatos estructurados.

**2. Relevancia**
El problema de deserción laboral (attrition) es un desafío crítico, común y medible en 
cualquier organización, lo que hace que los hallazgos sean directamente aplicables a 
contextos reales. Permite plantear hipótesis de alto valor como: *¿influye el equilibrio 
vida-trabajo en la renuncia de empleados?* o *¿existe relación entre el nivel salarial 
y la probabilidad de abandono?*. En contraste, el dataset de Stack Overflow es 
excesivamente grande y disperso para el alcance de este proyecto, y el de Reptiles, 
aunque técnicamente válido, aborda un dominio de nicho sin aplicación directa al 
análisis organizacional.

**3. Documentación**
Cuenta con un diccionario de datos completo que define los 35 atributos con claridad 
(edad, departamento, satisfacción laboral, nivel de cargo, etc.), además de alta 
actividad en Kaggle con múltiples notebooks de referencia. El dataset de Reptiles tiene 
buena organización visual pero carece de metadatos por imagen. El de Stack Overflow, 
aunque tiene documentación oficial robusta, presenta baja actividad en su versión de 
Kaggle.

**4. Manejabilidad**
Al ser tabular y de tamaño reducido (227.98 kB), es computacionalmente ligero y permite 
aplicar directamente técnicas de codificación de variables categóricas, escalado numérico 
y reducción de dimensionalidad mediante PCA, cubriendo los requerimientos técnicos de las 
fases siguientes. El dataset de Stack Overflow, con 94.84 MB y ~85 columnas, requeriría 
un esfuerzo de limpieza y selección de variables considerablemente mayor. El de imágenes 
exigiría infraestructura de GPU y arquitecturas de deep learning fuera del alcance 
de este proyecto. Además, vale la pena mencionar que tiene el mejor índice de manejabilidad 
en Kaggle. Un poco por encima del dataset de reptiles y superando por mucho al de Stack Overflow.
