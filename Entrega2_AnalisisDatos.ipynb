## Fase 1: Exploración de Bases de Datos

### Para esta fase, documentamos tres opciones de fuentes de datos de dos tipos diferentes (tabular, texto e imàgenes).

### 1. IBM HR Analytics Employee Attrition & Performance

Fuente y tipo: Secundaria (Kaggle), datos tabulares.

Características: 1,470 registros, 35 atributos (mezcla de variables numéricas y categóricas).

Documentación y aplicación: Muy alta. Ideal para predecir la rotación de empleados y optimizar la retención de talento.

Tamaño: 227.98 kB

### 2. Stack Overflow Annual Developer Survey 

Fuente y tipo: Secundaria (Kaggle/Stack Overflow), datos tabulares.

Características: Más de 80,000 registros, múltiples atributos sobre lenguajes (Java, JS, PHP, SQL) y entornos de bases de datos.

Documentación y aplicación: Baja, con pocas discusiones, una entrada de código. Se utiliza para análisis de mercado laboral tecnológico y tendencias de desarrollo.

Tamaño: 94.84 MB

### 3. Reptiles and Amphibians Image Dataset

Fuente y tipo: Secundaria (Kaggle), datos de imágenes (Computer Vision).

Características: Aproximadamente 6,045 imágenes en formato .jpg, organizadas en carpetas por clases (especies), con diversidad de resoluciones y entornos naturales.

Documentación y aplicación: Alta. Es ideal para el entrenamiento de redes neuronales convolucionales (CNN), aplicaciones de identificación taxonómica y proyectos de monitoreo de biodiversidad.

Tamaño: 141.05 MB


## Justificación

Sustentación de Selección: IBM HR Analytics
1. Credibilidad y Fuente
Origen: Es un dataset de tipo secundario, generado por científicos de datos de IBM para fines educativos y de práctica profesional.


2. Relevancia y Aplicación
Potencial de Hipótesis: Este dataset es ideal para cumplir con la Fase 2 , ya que nos permite plantear hipótesis de alto valor como: ¿Influye el equilibrio vida-trabajo en la renuncia de empleados? o "¿Existe una brecha salarial por género en departamentos específicos?.

- Comparativa con los otros 2 datasets: Mientras que el de Reptiles es de nicho y el de Stack Overflow es excesivamente grande y disperso, el de IBM trata sobre deserción laboral (attrition), un problema crítico, comun y universal en las organizaciones.

3. Documentación
Claridad: Posee un nivel de documentación alto, con un diccionario de datos que define claramente qué significa cada uno de los 35 atributos (edad, departamento, satisfacción laboral, etc.).

- Atributos: Cuenta con una mezcla equilibrada de datos para realizar un Análisis Univariado y Multivariado completo, permitiendo el uso de histogramas, boxplots y mapas de calor de correlación de forma efectiva como lo suguiere esta entrega.

4. Manejabilidad
Preprocesamiento: A diferencia del dataset de imágenes (Reptiles), este es de tipo tabular, lo que facilita la aplicación de codificación de variables categóricas (como el departamento) y escalado de variables numéricas (como el salario ).

- Reducción de Dimensionalidad: Al tener 35 columnas, es el candidato perfecto para aplicar PCA (Análisis de Componentes Principales), cumpliendo con aplicación  de requerimientos de la Fase 3 de la entrega.
